#Data pre-processing

```python
# Determine the number of missing values for every column.
Df.isnull().sum()





```

```python
# Replacing all NaN values in 'reviews_per_month' with zero.
Df.fillna({'reviews_per_month':0},inplace=True)
```

```python
# Removing  columns that are not required  from the Dataset .
Df.drop(['id','last_review','name'],axis=1,inplace=True)
```

```python
# Excluding property with zero listed price and zero availability.
Df = Df.loc[ Df['price'] & Df['availability_365'] > 0 ]
```

```python
Df.describe()
```

#Exploratory analysis and Visualisation

**1. Checking the distribution of Data.**

```python
Features = ['price','minimum_nights','calculated_host_listings_count','availability_365','number_of_reviews','reviews_per_month']
plt.figure(figsize=(15,10))

for i in enumerate(Features):
  plt.subplot(2,3,i[0]+1)
  sns.distplot(Df[i[1]])


```

* As we can see most of the features in our dataset are right skewed.
* Skewness helps in understanding the shape and outliers in a dataset.
* Depending on the model, skewness in the values of a specific independent variable (feature) may violate model assumptions.

**2.** **Which neighbourhood_group has the most number of listing ?**

```python
df = pd.DataFrame(Df.groupby(['neighbourhood_group','room_type'])['room_type'].count().unstack())
df
```

```python
# Creating a bar plot
df.plot(kind='bar')
plt.title('Room_type_listings')
plt.ylabel('no_of_listings')
plt.xlabel('neighbourhood_group')
plt.rcParams['figure.figsize'] = (7, 7)
```

1.Most number of listings are in Manhattan and the least number of listings are in Staten island.

2.Entire home\apt is the most listed property type and Shared room is the least listed property type.

3.Because less number of people prefer to stay in shared room type ,so it has the least number of listings.


**3. Which is the most listed property type?**

```python
df = pd.DataFrame(Df.groupby('room_type')['room_type'].count())
```

```python
# Creating pie chart.
fig = plt.figure(figsize =(10, 7))
plt.pie(df['room_type'], labels = df.index , autopct= '%1.2f%%')

# show plot
plt.show()
```

Entire home\apt is the most listed room_type and shared room is the least listed room_type.

**4. Price of different property types.**

```python
property_type = pd.DataFrame(Df.groupby('room_type')['price'].agg('mean'))
property_type
```

```python
property_type.plot(kind='bar',color=['b'])
plt.title('Property type_vs_ price')
plt.ylabel('price')
plt.xlabel('room_type')
plt.rcParams['figure.figsize'] = (5, 5)
```

Shared room type is the cheapest and entire home\apt is the most expensive property type.

**5. Making boxplots**

```python
Features = ['price','minimum_nights','calculated_host_listings_count','availability_365','number_of_reviews','reviews_per_month']
plt.figure(figsize=(15,10))

for i in enumerate(Features):
  plt.subplot(2,3,i[0]+1)
  sns.boxplot(x = Df[i[1]])

```

 **6. Which is the most expensive neighbourhood group ?**

```python
neighbourhood_group_df = pd.DataFrame(Df.groupby(['neighbourhood_group','room_type'])['price'].mean().unstack())
neighbourhood_group_df
```

```python
# Making a bar plot.
neighbourhood_group_df.plot(kind='bar')
plt.title('neighbourhood_vs_ price')
plt.ylabel('price')
plt.xlabel('neighbourhood_group')
plt.rcParams['figure.figsize'] = (5, 5)
```

1. Manhattan is the most expensive neighbourhood group while Brooklyn is the second most expensive.

2. Major cause for the maximum number of listings in these neighbourhood groups could be the high prices.


**7. Price relation with room type boxplot**

```python
# price per room type for properties under 175$.
title = 'properties under 175$'
price_data = Df.loc[Df['price'] < 175]
sns.set(style = 'whitegrid')
sns.boxplot(x= price_data['room_type'] , y= price_data['price'] , data= price_data)
plt.title(title)
plt.show()

```

```python
# price per room type for properties above 175$.
title = 'properties above 175$'
price_data = Df.loc[Df['price'] > 175]
sns.set(style = 'whitegrid')
sns.boxplot(x= price_data['room_type'] , y= price_data['price'] , data= price_data)
plt.title(title)
plt.show()
```

**8. Maximum number of nights stayed in which room type ?**

```python
df = pd.DataFrame(Df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].max().unstack())
df
```

```python
df.plot(kind = 'bar', width = 0.8)
plt.rcParams['figure.figsize'] = (10, 5)
plt.ylabel('number_of_nights')
plt.show()
```



1.  people usually prefer to spend less number of nights in shared room type. May be because of lack of privacy.

2.  Maximum number of nights spent is in Entire home/apt.
3.  Most people prefer to stay in entire homes or apartments.
4. In Manhattan no. of nights in shared room type is more than in private room type because of the high price.





**9. Which hosts are the busiest ?**

```python
Busiest_hosts = Df.groupby(['host_name','neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
Busiest_hosts = Busiest_hosts.sort_values(by = 'minimum_nights',ascending = False).head(5)
```

```python
Busiest_hosts
```

```python
name = Busiest_hosts['host_name']
night_stayed = Busiest_hosts['minimum_nights']
plt.rcParams['figure.figsize'] = (10, 5)
plt.bar(name,night_stayed,width = 0.5,color = 'red')
plt.xlabel('Name of the host')
plt.ylabel('Minimum number of nights stayed')
plt.title('Busiest hosts')
plt.show()
```

Busiest hosts are :

1. Sonder(NYC)
2. Blueground
3. Kara
4. Sonder
5. Ken

**12. Number of reviews vs price**

```python
fig = plt.figure(figsize=(10,5))
plt.scatter(Df['price'],Df['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.title('No_of_reviews vs price')
plt.show()
```

1. We can clearly see in the above scatterplot that maximum number of reviewed properties lies below the price  of 2000 USD.
2. properties above the price of 2000 USD gets less number of bookings , so the number of reviews are also low.

**13. What is the relationship between minimum number of nights and price?**

```python
fig = plt.figure(figsize=(10,5))
plt.scatter(Df['minimum_nights'],Df['price'])
plt.xlabel('minimum_nights')
plt.ylabel('price')
plt.title('minimum_nights vs price')
plt.show()
```

Seeing the  above scatterplot we can say that on an average in a  neighbourhood group for a given room type lesser the minimum number of nights more is its price .

**14. Average number of reviews vs Neighbourhood group**

```python
df = pd.DataFrame(Df.groupby(['neighbourhood_group'])['number_of_reviews'].mean())
df
df.plot(kind='bar')
plt.title('neighbourhood_vs_ reviews')
plt.ylabel('reviews')
plt.xlabel('neighbourhood_group')

```

1. Brookyln is the most reviewed neighbourhood group.
2. Bronx and Manhattan are the least reviewed neighbourhood groups.

 **15. Creating a heat map to see correlation between different variables.**

```python
Airbnb_data_df = Df[['room_type', 'price', 'minimum_nights', 'number_of_reviews',
       'reviews_per_month']]
```

```python
corr_df = Airbnb_data_df.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr_df , annot=True)
plt.show()
```
